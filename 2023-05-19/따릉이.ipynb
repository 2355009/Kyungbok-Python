```python
import numpy as np
np.sort(df['시간'].unique())
```

```python
df['시간'].apply(lambda x : '오후' if x<18 else '오전')
```

```python
df['시간대']=df['시간'].apply(lambda x : '오후' if x<18 else  
               ('밤' if 20<=x<24 else '야밤')
               )
df
```

```python
df['요일']=df['요일'].map(day_dict)
df
```
